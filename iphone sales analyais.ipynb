# What are the top 10 highest-rated iPohones on Flipkart?

```python
data.head()
```

```python
highest_rated = data.sort_values(by = ["Star Rating"], ascending = False )
highest_rated = highest_rated.head(10)
```

```python
print(highest_rated['Product Name'])
```

```python
print (highest_rated)
```

# How many ratings do the highest-rated iPhones on Flipkart have?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts() 
label = iphones.index 
counts = highest_rated['Number Of Ratings' ] 
figure = px.bar(highest_rated, x=label, y = counts)
figure.show()
```

# Which iPhone has the highest number of reviews on Flipkart?

```python
data.head()
```

```python
iphones = highest_rated['Product Name'].value_counts() 
label = iphones.index 
counts = highest_rated['Number Of Reviews'] 
figure = px.bar(highest_rated, x=label, y = counts, title = "number if reviews of Highest Rated iPhones")
figure.show()
```

# What is the relationship between the sale price of iPhones and the number of ratings on Flipkart? 

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings", y = "Sale Price", size = "Discount Percentage", trendline = 'ols', title = "Relationship between sale price and number of rating")
figure.show()
```

There is a negative linear relationship between the sale price of iPhones and the number of ratings. It means iPhones with lower sale prices are sold more.

# What is the relationship between the discount percentage and the number of ratings of iPhones on Flipkart? 

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings", y = "Discount Percentage", size = "Sale Price", trendline = 'ols', title = "Relationship between Discount and Number of Ratings")
figure.show()
```

# Can you figure out the least expensive and most expensive iPhones in the market, along with all their specifications?

```python
most_expensive = data.loc[data['Sale Price'].idxmax()]
least_expensive = data.loc[data['Sale Price'].idxmin()]

#Display the results
print ("Most Expensive Prodict:")
print(most_expensive)

print("\nLeast Expensive Product:")
print(least_expensive)
```
